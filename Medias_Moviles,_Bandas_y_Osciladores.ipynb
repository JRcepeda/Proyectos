```python
import yfinance as yf
import ta
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos=yf.Ticker('BVN').history(start='2020-07-01',end='2020-10-31')
```

```python
datos.head()
```

```python
datos['Close']=datos['Stock Splits']+datos['Dividends']+datos['Close']
```

```python
datos['SMA Short 5']=datos['Close'].rolling(window=5).mean()
datos['SMA Long 20']=datos['Close'].rolling(window=20).mean()
```

```python
datos.head(21)
```

```python
plt.figure(figsize=(16,8))
plt.plot(datos['Close'])
plt.plot(datos['SMA Short 5'],label='Media Movil 5')
plt.plot(datos['SMA Long 20'],label='Media Movil 20')
plt.title('Medias movil')
plt.legend()
plt.show()
```

```python
#media movil exponencial
datos['M Exp3']=datos['Close'].ewm(span=3,adjust=False).mean()
datos['M Exp5']=datos['Close'].ewm(span=5,adjust=False).mean()
datos['M Exp20']=datos['Close'].ewm(span=20,adjust=False).mean()
```

```python
bb=ta.volatility.BollingerBands(close=datos['Close'],window=20,window_dev=2)
```

```python
datos['bb mean']=bb.bollinger_mavg()
datos['bb up']=bb.bollinger_hband()
datos['bb down']=bb.bollinger_lband()
```

```python
datos.head()
```

```python
#Bandas de Bollinger
plt.figure(figsize=(16,8))
plt.title('Bandas de Bollinger')
plt.plot(datos['Close'])
plt.plot(datos['bb up'],label='Sup',alpha=0.4)
plt.plot(datos['bb mean'],label='Med',alpha=0.4)
plt.plot(datos['bb down'],label='Inf',alpha=0.4)
plt.grid()
plt.legend()
plt.show()
```

```python
#MACD media movil de convergencia / divergencia
datos['MACD']=ta.trend.MACD(close=datos['Close'],window_fast=12,window_slow=26).macd()
```

```python
plt.figure(figsize=(16,8))
plt.title('Indicador de Convergencia')
plt.plot(datos['MACD'],label='MACD')
plt.legend()
plt.axhline(y=0,color='black')
plt.show()
```
