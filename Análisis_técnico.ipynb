```python
import pandas as pd
import numpy as np
import yfinance as yf
import mplfinance as mp
import matplotlib.pyplot as plt
```

```python
cemex=yf.Ticker('CX').history(start='2020-01-01',end='2021-01-01')
ecopetrol=yf.Ticker('EC').history(start='2020-01-01',end='2021-01-01')
```

```python
cemex.describe()
```

```python
ecopetrol.describe()
```

```python
mp.plot(
    cemex,
    type='candle',
    style='yahoo',
    volume=True,
    title='Velas Japonesas Cemex',
    figratio=(15,8)
    )
```

```python
mp.plot(
    cemex['2020-03-01':'2020-04-01'],
    type='candle',
    style='yahoo',
    volume=True,
    title='Velas japonesas ECOPETROL',
    figratio=(15,8)
)
```

```python
mp.plot(
    ecopetrol['2020-03-01':'2020-04-01'],
    type='candle',
    style='yahoo',
    volume=True,
    title='Velas japonesas ECOPETROL',
    figratio=(15,8)
    #mav# permite agregar
)
```

```python
eco=yf.Ticker('EC').history(start='2022-03-13',end='2023-03-13',interval="1h")
```

```python
eco.head()
```

```python
mp.plot(
    eco,
    type='candle',
    style='yahoo',
    figratio=(15,8),
    volume=True,
    mav=(5,20),#dea movil de 20 horas
    title='Velas japonesas comp Hora'
)
```
